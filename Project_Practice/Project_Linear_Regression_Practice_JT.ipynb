```python
#Load data
TSLA_stock_hist_hourly_2021_11_7_df = pd.read_csv(Path('Resources/TSLA_stock_hist_hourly_2021_11_7.csv'))
TSLA_stock_hist_hourly_2021_11_7_df.head()
```

```python
# Create a grade level DataFrames.

tsla_percent = TSLA_stock_hist_hourly_2021_11_7_df[(TSLA_stock_hist_hourly_2021_11_7_df["Percent_Day_Change"] == "TSLA %")]

spy_percent = SPY_stock_hist_hourly_2021_11_7_df[(SPY_stock_hist_hourly_2021_11_7_df["Percent_Day_Change"] == "SPY %")]

tsla_tweet_count = TSLA_tweet_7days_prior_hist_2021_11_7_perhour_df[(TSLA_tweet_7days_prior_hist_2021_11_7_perhour_df["tweet_counts"] == "Tweets")]

date_hour = TSLA_tweet_7days_prior_hist_2021_11_7_perhour_df[(TSLA_tweet_7days_prior_hist_2021_11_7_perhour_df["date_UTC"] == "Date/Time")]
```

```python
# Combine percent change, tweets counted, and date by hour into one dataframe 
combo_df = pd.DataFrame({
               "Date by Hour": date_hour,
               "SPY %": spy_percent,
               "TSLA %": tsla_percent,
               "Tweet Count": tsla_tweet_count})

combo_df.head()
```
